```python
#REPLACING THE TARGET VALUE WITH PRICE
```

```python
df= pd.DataFrame(boston_datasets.data, columns=boston_datasets.feature_names)
df['Price']= boston_datasets.target
```

```python
df
```

```python
# SPLITTING THE DATA INTO TRAINING and EVALUATING DATA
```

```python
from sklearn.model_selection import train_test_split
x= df.loc[ :, df.columns != 'Price']#training data
y= df.loc[:, df.columns == 'Price']#testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=123)
```

```python
#DETERMINING THE INPUTS, HIDDEN LAYERS AND OUTPUTA
```

```python
from keras.models import Sequential
from keras.layers import Dense
model= Sequential()
model.add(Dense(128, input_shape=(13, ), activation= 'relu', name= 'dense_layer_1'))
model.add(Dense(64, activation= 'relu', name= 'dense_layer_2'))
model.add(Dense(1, activation='linear', name='dense_output'))
```

```python
#MODEL COMPILATION
```

```python
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
#TRAINING
```

```python
model.fit(x_train, y_train, epochs=100)
```

```python
#MODEL EVALUATION
```

```python
loss= model.evaluate(x_test, y_test)
print('Mean square error on test data:', loss)
```
